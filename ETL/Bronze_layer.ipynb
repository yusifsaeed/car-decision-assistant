```python
from pyspark.sql.functions import *

df_hatla2ee = (
    spark.read.format("csv")
    .option("header", "true")
    .option("inferSchema", "false") 
    .load("/Workspace/final project/hatla2ee.csv")
)

df_contact = (
    spark.read.format("csv")
    .option("header", "true")
    .option("inferSchema", "false") 
    .load("/Workspace/final project/Data_ContactCars.csv")
)
```

```python
df_hatla2ee.display()
df_contact.display()
```

```python
df_hatla2ee = df_hatla2ee.withColumnRenamed('Transmission type', 'Transmission')  
df_hatla2ee = df_hatla2ee.withColumnRenamed('Fuel type', 'Fuel_type') 
df_hatla2ee = df_hatla2ee.withColumnRenamed("Posted On", "Date_Posted")  
df_contact = df_contact.withColumnRenamed("Date Posted", "Date_Posted")
df_contact = df_contact.withColumnRenamed("Engine CC", "Engine_CC")
result = df_hatla2ee.unionByName(df_contact, allowMissingColumns=True) 
```

```python
result.display()
```

```python
spark.sql("DROP TABLE IF EXISTS final_project.Bronze_table")
result.write.mode("overwrite").format("delta").saveAsTable("final_project.Bronze_table")
```
